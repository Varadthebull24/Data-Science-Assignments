### 7) Building KNN Model

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['animal name']= label_encoder.fit_transform(df['animal name']) 
```

```python
x= df.iloc[:, 1:17]
y = df.iloc[:, -1]
x
```

```python
y
```

```python
model = KNeighborsClassifier(n_neighbors = 17)
```

```python
model.fit(x,y)
```

```python
kfold = KFold(n_splits=12)
results = cross_val_score(model, x, y, cv = kfold)
```

```python
results
```

```python
results.mean()
```

### 8) Tuning the Algorithm

```python
n_neighbors = list(range(1,40))
parameters = {'n_neighbors' : n_neighbors}
```

```python
model = KNeighborsClassifier()
grid = GridSearchCV(estimator = model, param_grid = parameters)
grid.fit(x, y)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

### 9) K-Score

```python
k_range = range(1,35)
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    scores = cross_val_score(knn, x, y, cv=4)
    k_scores.append(scores.mean())
```

```python
k_scores
```

```python
df_accuracy = pd.DataFrame({'Value of k' : range(1,35),'Accuracy' : k_scores})
df_accuracy
```

```python
df_accuracy.sort_values('Accuracy', ascending = False)
```
