```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("book.csv")
```

### 3) Performing Apriori Algorithm

##### Association rules with 10% Support and 70% confidence

```python
df1 = apriori(df, min_support=0.1, use_colnames=True)
df1
```

```python
rules = association_rules(df1, metric="lift", min_threshold=0.7)
rules
```

##### Note:- Consequent is highly dependent on the antecedent

```python
rules.sort_values('lift',ascending = False)[0:20]
```

```python
rules[rules.lift>1]
```

```python
plt.figure(figsize=(16,9))
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```

```python
support = rules["support"]
confidence =  rules["confidence"]
lift = rules["lift"]
```

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111, projection = '3d')
ax1.scatter(support,confidence,lift)
ax1.set_xlabel("support")
ax1.set_ylabel("confidence")
ax1.set_zlabel("lift");
```

```python
rules[['support','confidence','lift']].hist();
```

```python
plt.figure(figsize=(16,9),dpi=250)
sns.barplot(x="support",y="confidence",data=rules)
plt.title("support vs confidence")
plt.show()
```

```python
sns.scatterplot(data=rules, x="support", y="confidence", hue="antecedents")
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.show()
```
