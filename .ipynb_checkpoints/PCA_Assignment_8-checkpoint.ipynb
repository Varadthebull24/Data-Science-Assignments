### 7) Performing PCA

```python
data2=data1.values
data2
```

```python
data3=scale(data2)
data3
```

```python
pca= PCA(n_components = 13)
pca_values = pca.fit_transform(data3)
pca_values
```

```python
pca.components_
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 8)*100)
print("The amount of variance that each PCA explains is ",var1)
```

```python
plt.plot(var1,color="red");
```

```python
data4 = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3']), data['Type']], axis = 1)
data4
```

```python
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data = data4);
```

### 8) Performing Heirarchial Clustering

```python
plt.figure(figsize = (15, 8));
dendrogram = sch.dendrogram(sch.linkage(data3, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'average')
```

```python
y_hc = hc.fit_predict(data1)
Clusters = pd.DataFrame(y_hc,columns=['Clusters'])
Clusters.head()
```

```python
data5 = pd.concat([data1,Clusters],axis=1)
data5.head(7)
```

```python
data5['Clusters'].value_counts()
```

### 9) K-Means

```python
wcss = []
for i in range(1, 8):
    kmeans = KMeans(n_clusters=i,random_state=2)
    kmeans.fit(data1)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 8), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
Clusters2 = KMeans(3, random_state=27)
Clusters2.fit(data1)
```

```python
Clusters2.labels_
```

```python
data6 = pd.DataFrame(Clusters2.labels_, columns =['Clusters'])
```

```python
data7 = pd.concat([data1,data6], axis=1)
data7
```

```python
data7.groupby('Clusters').agg(['mean'])
```
