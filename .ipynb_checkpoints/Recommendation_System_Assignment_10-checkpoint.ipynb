### 8) Sorting User ID

```python
df2 = df1.pivot_table(index='UserId', columns='Title', values='Rating')
```

```python
df2
```

##### Note:- We weill have to Impute those NaN's with 0 values

```python
df2.fillna(0, inplace=True)
```

```python
df2.index = df1['UserId'].unique()
df2
```

### 9) Let's Calculate Cosine Similarity between Users`

```python
user1 = 1 - pairwise_distances( df2.values, metric='cosine')
user1
```

```python
np.fill_diagonal(user1, 0)
```

```python
user = pd.DataFrame(user1)
user
```

### 11) We will Consider most Similar Customer

```python
def recommend(custID):
    simID = user.loc[custID].idxmax()
    simID_books = df1[df1['UserId'] == simID].Title
    custID_books = df1[df1['UserId'] == custID].Title
    return set(simID_books) - set(custID_books)
```

```python
recommend(8)
```

```python
recommend(3462)
```

```python
recommend(162113)
```
